```python
import csv

from google.colab import files
uploaded = files.upload()

from sys import float_repr_style
def punteggio_pilota():

  Nome_pilota=str(input("inserire Cognome pilota con lettera iniziale maiuscola: "))
  with open("formula1_data.csv") as file_piloti:

    csv_reader = csv.DictReader(file_piloti)

    pilota_stats=[]
    total_score=0
    total_wins=0
    total_podium=0
    for row in csv_reader:
        if row["Driver"] == Nome_pilota:
                Positioning = int(row["Position"])
                if Positioning == 1:
                  total_score=total_score+10
                  total_wins=total_wins+1
                  total_podium=total_podium+1
                elif Positioning == 2:
                  total_score=total_score+8
                  total_podium=total_podium+1
                elif Positioning == 3:
                  total_score=total_score+6
                  total_podium=total_podium+1
                elif Positioning == 4:
                  total_score=total_score+5
                elif Positioning == 5:
                  total_score=total_score+4
                elif Positioning == 6:
                  total_score=total_score+3
                elif Positioning == 7:
                  total_score=total_score+2
                elif Positioning == 8:
                  total_score=total_score+1
                else:
                  total_score=total_score+0
  pilota_stats.append(total_score)
  pilota_stats.append(total_wins)
  pilota_stats.append(total_podium)
  return pilota_stats



def classifica_piloti():
  dict_piloti={}
  with open("formula1_data.csv") as file_piloti:
    csv_reader = csv.DictReader(file_piloti)

    for row in csv_reader:
        driver =row["Driver"]
        Posizione = int(row["Position"])

        punteggio_tot= dict_piloti.get(driver,0)
        if Posizione == 1:
                  punteggio_tot= punteggio_tot+10
        elif Posizione == 2:
                  punteggio_tot=punteggio_tot+8
        elif Posizione == 3:
                  punteggio_tot=punteggio_tot+6
        elif Posizione == 4:
                  punteggio_tot=punteggio_tot+5
        elif Posizione == 5:
                  punteggio_tot=punteggio_tot+4
        elif Posizione == 6:
                  punteggio_tot=punteggio_tot+3
        elif Posizione == 7:
                  punteggio_tot=punteggio_tot+2
        elif Posizione == 8:
                  punteggio_tot=punteggio_tot+1
        else:
                  punteggio_tot=punteggio_tot+0

        dict_piloti[row["Driver"]]=punteggio_tot

        dict_piloti = dict(sorted(dict_piloti.items(), key=lambda item: item[1], reverse=True))
    return dict_piloti


with open("Drivers_Standings_2008.txt", "w") as Drivers_Standings_2008:
  for k,v in classifica_piloti().items():
    Drivers_Standings_2008.write(f"{k} : {v}\n")


def classifica_costruttori():
  dict_team={}
  dict_teams_driver={}
  dict_piloti=classifica_piloti()
  with open("formula1_data.csv") as file_piloti:
    csv_reader = csv.DictReader(file_piloti)

    for row in csv_reader:
        team =row["Team"]
        driver =row["Driver"]
        dict_teams_driver[driver]=team

  with open("Drivers_Standings_2008.txt", "r") as Drivers_Standings_2008:
    for line in Drivers_Standings_2008:
      driver, score = line.split(":")
      driver = driver.strip()
      score = int(score.strip())
      team=dict_teams_driver.get(driver)
      if team:
        dict_team[team] = dict_team.get(team, 0) + score

  dict_team = dict(sorted(dict_team.items(), key=lambda item: item[1], reverse=True))
  return dict_team













```
