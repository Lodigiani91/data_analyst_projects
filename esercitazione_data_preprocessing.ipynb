```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ProfAI/machine-learning-fondamenti/main/datasets/housing_dirty.csv",index_col=0)
df.describe()
```

```python

df.isna().sum()
```

```python
df.count()<df.shape[0]*0.7
df= df.drop("LSTAT",axis=1)
df.head(9)
```

```python
## rimozione righe con valori mancanti > del 25%

row_na=df.isna().sum(axis=1)
row_na = row_na[row_na>df.shape[1]*0.25]
row_na
to_remove = row_na.index.values
df = df.drop(to_remove)
df.shape

## oppure

df = df.dropna(thresh=df.shape[1]*0.75)
df.shape
```

```python
row_price_na = df["PRICE"].isna()
row_price_na = row_price_na[row_price_na>0]
row_price_na

##oppure
## df = df.dropna(subset=["PRICE"])
## df.shape
```

```python
to_remove = row_price_na.index.values
df = df.drop(to_remove)
df.shape
```

```python
cols= df.columns

cols=cols.drop(["CRIM", "CHAS"])
cols
```

```python
replace_with_mean = df[cols].mean().round(2)
df[cols]=df[cols].fillna(replace_with_mean)
na_count = df[cols].isna().sum().sum()
na_count
```
